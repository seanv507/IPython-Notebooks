```python
def principal_remaining(nper, n_payments, period_rate, amount):
    q=(1+period_rate/100.0)
    q_d=q**nper
    principal_remaining=amount*(q_d - q**n_payments)/(q_d-1)
    return principal_remaining
```

```python
def pmt(start_date,nper, period_rate, amount):
    index=pd.bdate_range(start_date, periods=nper+1, freq='1m') # this does end of month -grrr!!!
    period_rate/=100.0
    q=(1+period_rate)
    q_d=q**nper
    installment=amount*period_rate * q_d/(q_d -1)
    princ=amount*(q_d - q**np.arange(nper+1))/(q_d-1) # from start (full principal) to 0)
    cashflows=pd.DataFrame({'principal':princ, 'installment':installment },index=index)
    cashflows['interest_payment']=cashflows.principal.shift(1)*period_rate
    cashflows['principal_payment']=cashflows.principal.shift(1)-cashflows.principal
    return cashflows
```

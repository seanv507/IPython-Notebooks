# Financial Functions

```python
def principal_remaining(nper, n_payments, period_rate, amount):
    q=(1+period_rate/100.0)
    q_d=q**nper
    principal_remaining=amount*(q_d - q**n_payments)/(q_d-1)
    return principal_remaining
```

```python
def gen_dates(FirstPayDate,RollDay, nper):
    pay_dates=np.datetime64(FirstPayDate,'M')+np.timedelta64(1,'M')*np.arange(nper)+np.timedelta64(RollDay-1,'D')
    
    # pay_dates=RollPayDate + pd.to_timedelta(np.arange(nper+1),unit='M')
    rolled_pay_dates= np.busday_offset(pay_dates,0,roll='forward')

    return rolled_pay_dates
```

```python
def pmt(LoanStartDate,FirstPaymentDate, RollDay,nper, period_rate, amount):
    pay_dates=gen_dates(FirstPaymentDate,RollDay, nper)
    dates=np.insert(pay_dates, 0, LoanStartDate)
    period_rate/=100.0
    q=(1+period_rate)
    q_d=q**nper
    instalment=amount*period_rate * q_d/(q_d -1)
    princ=amount*(q_d - q**np.arange(nper+1))/(q_d-1) # from start (full principal) to 0)
    cashflows=pd.DataFrame({'principal':princ, 'instalment':instalment },index=dates)
    cashflows.index.name='date'
    cashflows['interest_payment']=cashflows.principal.shift(1)*period_rate
    cashflows['principal_payment']=cashflows.principal.shift(1)-cashflows.principal
    return cashflows
```

```python
def xnpv(rs,amounts, taus):
    """ should work for scalars and vectors. 
    taus: is the day count fraction as of the valuation date (tau=0)
    """
    qs=np.power(1+rs ,taus)
    return np.sum(amounts*qs)
```

```python
def xirr(amounts,taus, guess=0.1):
    return scipy.optimize.newton(lambda r: xnpv(r,amounts, taus),guess)
```

```python
def calc_expected_payment(l, repayments):
    
    expected_cashflow=pmt(l.LoanDate.date(),np.datetime64(l.FirstPaymentDate.date()),int(l.MonthlyPaymentDay), int(l.LoanDuration),float(l.IssuedInterest)/12.0, float(l.FundedAmount))
    expected_cashflow['cum_principal_payment']=expected_cashflow.principal_payment.cumsum()
    ext_dates=gen_dates(l.FirstPaymentDate.date(),int(l.MonthlyPaymentDay), int(l.LoanDuration)+3)
    ext_dates=np.insert(ext_dates,0,l.LoanDate.date())
    # ugly!! 
    expected_cashflow=expected_cashflow.reindex(ext_dates )
    expected_cashflow['default_level']=expected_cashflow.cum_principal_payment.shift(3)
    expected_cashflow=expected_cashflow[:l.ReportAsOfEOD]
    cum_repayments=repayments[repayments.loan_id==l.name].sort_index()[[ u'PrincipalRepayment', u'InterestRepayment', u'LateFeesRepayment']].cumsum()
    # take actual cum repayments, move them back 5 days, and then assign to the next 'expected payment date'
    expected_actual=expected_cashflow.join(cum_repayments.tshift(-2,'D').reindex(expected_cashflow.index,method='ffill'))
    expected_actual['loan_id']=l.name
    # replace nans with 0 for comparison to be made even when no repayment.
    # add 5 euros to allow for any error in calculation 
    defaults=expected_actual[expected_actual.PrincipalRepayment.fillna(0)-1<=expected_actual.default_level]
    return expected_actual, defaults
```
